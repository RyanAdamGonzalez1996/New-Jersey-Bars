```python
# Setup all Dependencies
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time
import requests
```

```python
# Setup splinter

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
new_brunswick_name_list = []
new_brunswick_type_list = []
new_brunswick_website_list = []
new_brunswick_phone_list = []
new_brunswick_price_list = []
new_brunswick_address_list = []
new_brunswick_rating_list = []
new_brunswick_hours_list = []
new_brunswick_second_url = []

    
    
    
    
    
new_brunswick_url = "https://www.tripadvisor.com/Restaurants-g46664-New_Brunswick_New_Jersey.html"
browser.visit(new_brunswick_url)

#browser.find_by_name('send').first.click()

#Bar Name
for x in range(10):
    
    html = browser.html
    soup = BeautifulSoup(html,"html.parser")
    
    names = soup.find_all('div',class_= "wQjYiB7z")
   
    for result in names:
        try:
            restaurant = result.span.a.text
            link = "https://www.tripadvisor.com" + result.span.a['href']
            print(restaurant)
        except AttributeError as e:
            print(e)
            restaurant = ""
            link = ""
        new_brunswick_name_list.append(restaurant)
        new_brunswick_second_url.append(link)
            
    
                

#Bar Price                
    cuisine_price = soup.find_all('div',class_= "MIajtJFg _1cBs8huC _3d9EnJpt")
    for result in cuisine_price:
        try:
            t = result.find_all('span', class_ = 'EHA742uW')
            if len(t)>1:
                print(len(t))
                price_level = t[1].find('span', class_ = '_1p0FLy4t').text
            else:
                price_level = ''
            print(price_level)
        except AttributeError as e:
            print(e)
            price_level = ''
            
        new_brunswick_price_list.append(price_level)
       
    try:
        
        browser.click_link_by_partial_text('Next')
        time.sleep(1)
        html = browser.html
        
        
    except:
        print("Done")
#Bar Type
    cuisine = soup.find_all('div',class_= "MIajtJFg _1cBs8huC _3d9EnJpt")
    for result in cuisine:
        try:
            cuisine_type = result.span.span.text
            print(cuisine_type)
        except AttributeError as e:
            cuisine_type = ""
        new_brunswick_type_list.append(cuisine_type)

        
```

```python
for url in new_brunswick_second_url:
    response = requests.get(url)
    soup_2 = BeautifulSoup(response.text,"html.parser")
#Extract Address
    raw_address = soup_2.find_all('span', class_ ='_2saB_OSe')   
    try:
        if len(raw_address) > 0:
            address = raw_address[0].text
        
        else:
            address = ""
            print(address)
        address = raw_address[0].text
        
        print(address)
    except AttributeError as e:
        print(e)
        address = ""
        

        
        


#Extract Review
    new_review = soup_2.find('svg', class_ = "_3KcXyP0F")['title']
    try:
        review = new_review.split(' ')
        if len(review) > 0:
            review_number = review[0]
        else:
            review_number = ""
        print(review_number)
    except AttributeError as e:
        print(e)
        review_number = ''
    new_brunswick_rating_list.append(review_number)
    new_brunswick_address_list.append(address)

browser.quit()
```

```python
overall = {'restaurant_name' : new_brunswick_name_list,
           'bar_type': new_brunswick_type_list,
           'price_level': new_brunswick_price_list,
           'rating_list': new_brunswick_rating_list,
          
           'address_list':new_brunswick_address_list }




new_brunswick_raw_df = pd.DataFrame.from_dict(overall)
new_brunswick_df = new_brunswick_raw_df.drop_duplicates()
new_brunswick_df

```

```python
new_brunswick_bar_and_pub_df = new_brunswick_df.loc[new_brunswick_df['bar_type'] == "Bar, Pub"]
new_brunswick_american_df = new_brunswick_df.loc[new_brunswick_df['bar_type'] == "American, Bar"]
new_brunswick_irish_df = new_brunswick_df.loc[new_brunswick_df['bar_type'] == "Irish, Bar"]
new_brunswick_italian_df = new_brunswick_df.loc[new_brunswick_df['bar_type'] == "Italian, Bar"]
new_brunswick_bar_df = new_brunswick_df.loc[new_brunswick_df['bar_type'] == "Bar"]
new_brunswick_pub_df = new_brunswick_df.loc[new_brunswick_df['bar_type'] == "Pub"]
new_brunswick_seafood_df = new_brunswick_df.loc[new_brunswick_df['bar_type'] == "Bar, Seafood"]
new_brunswick_bar_and_american_df = new_brunswick_df.loc[new_brunswick_df['bar_type'] == "Bar, American"]

new_brunswick_final_df= pd.concat([new_brunswick_bar_and_pub_df,new_brunswick_american_df,new_brunswick_irish_df,new_brunswick_italian_df,new_brunswick_bar_df,new_brunswick_pub_df,new_brunswick_seafood_df,new_brunswick_bar_and_american_df], axis=0) 
new_brunswick_final_df
```
