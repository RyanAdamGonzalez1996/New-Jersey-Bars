```python
import pandas as pd
```

```python
data = "../Resources/atlantic_city_data.csv"
df = pd.read_csv(data)
df
```

```python
bar_and_pub_df = df.loc[df['bar_type'] == "Bar, Pub"]
american_df = df.loc[df['bar_type'] == "American, Bar"]
irish_df = df.loc[df['bar_type'] == "Irish, Bar"]
italian_df = df.loc[df['bar_type'] == "Italian, Bar"]
bar_df = df.loc[df['bar_type'] == "Bar"]
pub_df = df.loc[df['bar_type'] == "Pub"]
seafood_df = df.loc[df['bar_type'] == "Bar, Seafood"]
bar_and_american_df = df.loc[df['bar_type'] == "Bar, American"]

ac_df= pd.concat([bar_and_pub_df,american_df,irish_df,italian_df,bar_df,pub_df,seafood_df,bar_and_american_df], axis=0) 
ac_df
```

```python
ac_df.to_csv("../Resources/Alantic_City_data.csv",index=False)
```
