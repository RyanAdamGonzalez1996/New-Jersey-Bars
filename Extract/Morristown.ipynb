```python
# Setup all Dependencies
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time
import requests
```

```python
# Setup splinter

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

```

```python
morristown_name_list = []
morristown_type_list = []
morristown_website_list = []
morristown_phone_list = []
morristown_price_list = []
morristown_address_list = []
morristown_rating_list = []
morristown_hours_list = []
morristown_second_url = []

    
    
    
    
    
morristown_url = "https://www.tripadvisor.com/Restaurants-g60906-Morristown_Morris_County_New_Jersey.html"
browser.visit(morristown_url)

#browser.find_by_name('send').first.click()

#Bar Name
for x in range(10):
    
    html = browser.html
    soup = BeautifulSoup(html,"html.parser")
    
    names = soup.find_all('div',class_= "wQjYiB7z")
   
    for result in names:
        try:
            restaurant = result.span.a.text
            link = "https://www.tripadvisor.com" + result.span.a['href']
            print(restaurant)
        except AttributeError as e:
            print(e)
            restaurant = ""
            link = ""
        morristown_name_list.append(restaurant)
        morristown_second_url.append(link)
            
    
                

#Bar Price                
    cuisine_price = soup.find_all('div',class_= "MIajtJFg _1cBs8huC _3d9EnJpt")
    for result in cuisine_price:
        try:
            t = result.find_all('span', class_ = 'EHA742uW')
            if len(t)>1:
                print(len(t))
                price_level = t[1].find('span', class_ = '_1p0FLy4t').text
            else:
                price_level = ''
            print(price_level)
        except AttributeError as e:
            print(e)
            price_level = ''
            
        morristown_price_list.append(price_level)
       
    try:
        
        browser.click_link_by_partial_text('Next')
        time.sleep(1)
        html = browser.html
        
        
    except:
        print("Done")
#Bar Type
    cuisine = soup.find_all('div',class_= "MIajtJFg _1cBs8huC _3d9EnJpt")
    for result in cuisine:
        try:
            cuisine_type = result.span.span.text
            print(cuisine_type)
        except AttributeError as e:
            cuisine_type = ""
        morristown_type_list.append(cuisine_type)

        
```

```python
for url in morristown_second_url:
    response = requests.get(url)
    soup_2 = BeautifulSoup(response.text,"html.parser")
#Extract Address
    raw_address = soup_2.find_all('span', class_ ='_2saB_OSe')   
    try:
        if len(raw_address) > 0:
            address = raw_address[0].text
        
        else:
            address = ""
            print(address)
        address = raw_address[0].text
        
        print(address)
    except AttributeError as e:
        print(e)
        address = ""
        

        
        


#Extract Review
    new_review = soup_2.find('svg', class_ = "_3KcXyP0F")['title']
    try:
        review = new_review.split(' ')
        if len(review) > 0:
            review_number = review[0]
        else:
            review_number = ""
        print(review_number)
    except AttributeError as e:
        print(e)
        review_number = ''
    morristown_rating_list.append(review_number)
    morristown_address_list.append(address)

browser.quit()
```

```python
overall = {'restaurant_name' : morristown_name_list,
           'bar_type': morristown_type_list,
           'price_level': morristown_price_list,
           'rating_list': morristown_rating_list,
          
           'address_list':morristown_address_list }




morristown_raw_df = pd.DataFrame.from_dict(overall)
morristown_df = morristown_raw_df.drop_duplicates()
morristown_df

```
