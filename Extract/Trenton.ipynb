```python
# Setup all Dependencies
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time
import requests
```

```python
# Setup splinter

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
trenton_name_list = []
trenton_type_list = []
trenton_website_list = []
trenton_phone_list = []
trenton_price_list = []
trenton_address_list = []
trenton_rating_list = []
trenton_hours_list = []
trenton_second_url = []

    
    
    
    
    
trenton_url = "https://www.tripadvisor.com/Restaurants-g46874-Trenton_New_Jersey.html"
browser.visit(trenton_url)

#browser.find_by_name('send').first.click()

#Bar Name
for x in range(10):
    
    html = browser.html
    soup = BeautifulSoup(html,"html.parser")
    
    names = soup.find_all('div',class_= "wQjYiB7z")
   
    for result in names:
        try:
            restaurant = result.span.a.text
            link = "https://www.tripadvisor.com" + result.span.a['href']
            print(restaurant)
        except AttributeError as e:
            print(e)
            restaurant = ""
            link = ""
        trenton_name_list.append(restaurant)
        trenton_second_url.append(link)
            
    
                

#Bar Price                
    cuisine_price = soup.find_all('div',class_= "MIajtJFg _1cBs8huC _3d9EnJpt")
    for result in cuisine_price:
        try:
            t = result.find_all('span', class_ = 'EHA742uW')
            if len(t)>1:
                print(len(t))
                price_level = t[1].find('span', class_ = '_1p0FLy4t').text
            else:
                price_level = ''
            print(price_level)
        except AttributeError as e:
            print(e)
            price_level = ''
            
        trenton_price_list.append(price_level)
       
    try:
        
        browser.click_link_by_partial_text('Next')
        time.sleep(1)
        html = browser.html
        
        
    except:
        print("Done")
#Bar Type
    cuisine = soup.find_all('div',class_= "MIajtJFg _1cBs8huC _3d9EnJpt")
    for result in cuisine:
        try:
            cuisine_type = result.span.span.text
            print(cuisine_type)
        except AttributeError as e:
            cuisine_type = ""
        trenton_type_list.append(cuisine_type)

        
```

```python
for url in trenton_second_url:
    response = requests.get(url)
    soup_2 = BeautifulSoup(response.text,"html.parser")
#Extract Address
    raw_address = soup_2.find_all('span', class_ ='_2saB_OSe')   
    try:
        if len(raw_address) > 0:
            address = raw_address[0].text
        
        else:
            address = ""
            print(address)
        address = raw_address[0].text
        
        print(address)
    except AttributeError as e:
        print(e)
        address = ""
        

        
        


#Extract Review
    new_review = soup_2.find('svg', class_ = "_3KcXyP0F")['title']
    try:
        review = new_review.split(' ')
        if len(review) > 0:
            review_number = review[0]
        else:
            review_number = ""
        print(review_number)
    except AttributeError as e:
        print(e)
        review_number = ''
    trenton_rating_list.append(review_number)
    trenton_address_list.append(address)

browser.quit()
```

```python
overall = {'restaurant_name' : trenton_name_list,
           'bar_type': trenton_type_list,
           'price_level': trenton_price_list,
           'rating_list': trenton_rating_list,
          
           'address_list':trenton_address_list }




trenton_raw_df = pd.DataFrame.from_dict(overall)
trenton_df = trenton_raw_df.drop_duplicates()
trenton_df
```

```python
trenton_bar_and_pub_df = trenton_df.loc[trenton_df['bar_type'] == "Bar, Pub"]
trenton_american_df = trenton_df.loc[trenton_df['bar_type'] == "American, Bar"]
trenton_irish_df = trenton_df.loc[trenton_df['bar_type'] == "Irish, Bar"]
trenton_italian_df = trenton_df.loc[trenton_df['bar_type'] == "Italian, Bar"]
trenton_bar_df = trenton_df.loc[trenton_df['bar_type'] == "Bar"]
trenton_pub_df = trenton_df.loc[trenton_df['bar_type'] == "Pub"]
trenton_seafood_df = trenton_df.loc[trenton_df['bar_type'] == "Bar, Seafood"]
trenton_bar_and_american_df = trenton_df.loc[trenton_df['bar_type'] == "Bar, American"]

trenton_final_df= pd.concat([trenton_bar_and_pub_df,trenton_american_df,trenton_irish_df,trenton_italian_df,trenton_bar_df,trenton_pub_df,trenton_seafood_df,trenton_bar_and_american_df], axis=0) 
trenton_final_df
```
